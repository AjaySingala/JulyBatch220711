```python
import pandas as pd
import matplotlib.pyplot as plt

# The recommended way to enable inline plotting in the IPython Notebook.
%matplotlib inline
```

```python
data = pd.read_csv('air_quality.csv', index_col=0, parse_dates=True)
```

```python
data.head()
```

### Simple plot

```python
data.plot()
```

### Isolated plot

```python
data['station_paris'].plot()
```

### Scatter plot

```python
# Here alpha denotes the transparency factor.
data.plot.scatter(x='station_london', y='station_paris', alpha=.25)
```

### Box plot

```python
data.plot.box()
```

### Area plot

```python
# figsize specifies the size of the figure object.
# Subplots: make separate subplots for each column.
data.plot.area(figsize=(12,4), subplots=True)
```

### Bar plot

```python
data.iloc[:15].plot.bar()
```

### Histograms

```python
data['station_paris'].plot.hist()
```

```python
data.plot.hist(alpha=.5)
```

### Pie

```python
data.mean().plot.pie()
```

### Scatter matrix plot

```python
from pandas.plotting import scatter_matrix
```

```python
# figsize specifies the size of the figure object.
scatter_matrix(data, alpha=0.2, figsize=(6,6))
```

### Secondary y-axis

```python
# cumsum(): Cumulative Sum value over an axis.
data['station_london'].plot()
data['station_paris'].cumsum().plot(secondary_y=True)
#data['station_paris'].cumsum().plot()
```
